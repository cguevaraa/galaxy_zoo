```python
import pandas as pd
import random
from matplotlib import pyplot as plt
from PIL import Image
import os
```

```python
solution = pd.read_csv('training_solutions_rev1.csv')
```

```python
solution
```

```python
sample1 = solution.sort_values(by='Class1.1', ascending=False).head(100).copy()
sample2 = solution.sort_values(by='Class1.2', ascending=False).head(100).copy()
sample3 = solution.sort_values(by='Class1.3', ascending=False).head(100).copy()
```

```python
img_path = "images_training_rev1"
files = os.listdir(img_path)
```

```python
fig = plt.figure(figsize=(15,18))
    
for i in range(10):
    fig.add_subplot(6,5,i+1)
    n = random.randint(0,100)
    plt.imshow(Image.open(img_path+'/'+str(sample1.GalaxyID.values[n])+'.jpg'))
for i in range(10):
    fig.add_subplot(6,5,i+11)
    n = random.randint(0,100)
    plt.imshow(Image.open(img_path+'/'+str(sample2.GalaxyID.values[n])+'.jpg'))
for i in range(10):
    fig.add_subplot(6,5,i+21)
    n = random.randint(0,100)
    plt.imshow(Image.open(img_path+'/'+str(sample3.GalaxyID.values[n])+'.jpg'))
```
